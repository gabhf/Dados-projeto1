```python
%matplotlib inline

import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

print('Esperamos trabalhar no diretório')
print(os.getcwd())
```

```python
# mortalidade infantil por 1000 nascidos :
mi = pd.read_excel('child_mortality_0_5_year_olds_dying_per_1000_born.xlsx')
#mortalidade infantil como medidor para saúde e saneamento

#crescimento anual do PIB (vai até 2013)
grwth = pd.read_excel('gdp_total_yearly_growth.xlsx')
```

```python
growth = grwth.drop([97])
```

```python
afdb = ['Angola','Madagascar','Algeria','Benin','Botswana','Chad','Zambia','Egypt','Eritrea','Guinea','Kenya','Liberia','Morocco','Mozambique','Niger','Senegal','Uganda','Gabon','Cameroon',"Cote d'Ivoire",'South Africa','Ghana','Ethiopia','Nigeria','Tunisia','Equatorial Guinea','Mauritius','Seychelles','Namibia','Burkina Faso','Djibouti','Sao Tome and Principe','Sierra Leone','Somalia']
```

```python
sel_paises = mi.loc[mi['country'].isin(afdb)]
dados1 = sel_paises.set_index('country')
#dados1 se refere à primeira variável estudada (mortalidade infantil)

sele_paises = growth.loc[growth['country'].isin(afdb)]
dados3 = sele_paises.set_index('country')

#dados3 se refere à terceira variavel: crascimento do pib
```

```python
anos = list(range(1993,2010))

data1 = dados1.loc[:,anos]
y = data1.mean()
x = y.index

data3 = dados3.loc[:,anos]
i = data3.mean()
j = i.index

# Definicao dos objetos figura e eixo       
fig, ax1 = plt.subplots()

ax1.plot(x,y, 'b-')
# dá nome ao eixo x
ax1.set_xlabel('Tempo (anos)')
# dá nome ao eixo y1
ax1.set_ylabel('Média de mortalidade infantil nos países do AfBD', color='b')
# seta os limites dos eixos x e y
ax1.axis([1995,2009, 50,160])
# faz a cor das marcas no eixo y1 iguais a linha
for tl in ax1.get_yticklabels():
    tl.set_color('b')    
# Cria um eixo com valor de y independente, mas com o mesmo x
ax2 = ax1.twinx()

ax2.plot(j, i, 'r')
# dá nome ao eixo y2
ax2.set_ylabel('Média de crescimento anual do PIB dos países do AfBD', color='r')
# seta os limites dos eixos x e y
ax2.axis([1995,2009,0,12])
# faz a cor das marcas no eixo iguais a linha
for tl in ax2.get_yticklabels():
    tl.set_color('r')
plt.grid(True)
plt.show()  
```

Os anos de 1996 e 2006 foram marcantes pois representaram, respectivamente o quase colapso e um sinal de recuperação do Banco.
Fica muito claro na curva da taxa de crescimento do PIB que após 1996 o índice decai em contraponto à forte ascensão que vinha dos anos anteriores. Depois disso, de 2006 a 2008, esse índice se mantém constante após uma recuperação vinda dos anos anteriores.
Por conta de a mortalidade infantil ser um índice que varia menos no tempo este gráfico não é bom para se analisar.

###### A seguir, serão analisados diferentes tipos de gráficos para os anos de 2003 a 2010 para analisar a recuperação do bloco após a crise.

```python
mi_sel = mi.set_index('country')
#foram selecionados alguns anos perto de 1996, 2006 e 2018 para analisar
mi_anos = mi_sel.loc[:,[1992,1996,2000,2003,2006,2010,2014,2018]]
mi_anos.loc[:,"AfDB"] = "Não"
mi_anos.loc[afdb,"AfDB"] = "Sim"
mi_anos.hist (column=2003, by="AfDB")
mi_anos.hist (column=2010, by="AfDB")
```

A primeira série de gráficos refere-se à taxa de mortalidade infantil no ano de 2003 e a segunda no ano de 2010. Os histogramas mostram que na média geral mundial o índice melhorou, ou seja, diminuiu porém numa velocidade menor. Já na média dos países-membros, houve uma melhora mais perceptível até o ano de 2010. 

```python
growth_sel = growth.set_index('country')
growth_anos = growth_sel.loc[:,[1992,1996,2000,2003,2006,2010,2014,2018]]
growth_anos.loc[:,"AfDB"] = "Não"
growth_anos.loc[afdb,"AfDB"] = "Sim"
growth_anos.hist (column=2003, by="AfDB")
growth_anos.hist (column=2010, by="AfDB")
```

A primeira série de gráficos refere-se à taxa de crescimento anual do PIB no ano de 2003 e a segunda no ano de 2010. 

```python
mi_anos.boxplot (column=2003, by="AfDB")
mi_anos.boxplot (column=2010, by="AfDB")
```

O boxplot acima é referente à mortalidade infantil. É simples de se perceber que há uma variação maior para os países membros do AfBD em 2010 em relação a 2003 e que no mesmo período a diminuição foi bem mais estável no restante do mundo.

```python
growth_anos.boxplot (column=2003, by="AfDB")
growth_anos.boxplot (column=2010, by="AfDB")
```

Esse boxplot se refere ao crescimento anual do PIB. A análise dos gráficos demonstra uma melhora em 2010 em relação a 2003 nos países-membros ligeiramente mais expressiva que a média mundial.

## Conclusão

##### Os dados coletados mostram que após a crise do bloco em 1996 e antes de sua recuperação em 2006, a taxa de mortalidade infantil decresceu menos rapidamente se comparado aos anos ao redor da recuperação do Banco em 2006. Comportamento esse, que pode ser observado também na análise do crescimento do PIB desses países, que sofreu uma queda após a crise e mostrou recuperação de 2004 a 2008.
 ##### Além disso, de 1996 a 2006, as médias mundiais desses indicadores mostraram-se muito mais estáveis.
