```python
%matplotlib inline

import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

print('Esperamos trabalhar no diretório')
print(os.getcwd())
```

```python
# mortalidade infantil por 1000 nascidos :
mi = pd.read_excel('child_mortality_0_5_year_olds_dying_per_1000_born.xlsx')
#porcentagem de adultos (15 a 49) com hiv estimada:
hiv = pd.read_excel('newly_hiv_infected_percent_age_15_49.xlsx')
mi
```

```python
afdb = ['Angola','Madagascar','Algeria','Benin','Botswana','Chad','Zambia','Egypt','Eritrea','Guinea','Kenya','Liberia','Morocco','Mozambique','Niger','Senegal','Uganda','Gabon','Cameroon',"Cote d'Ivoire",'South Africa','Ghana','Ethiopia','Nigeria','Tunisia']
grupo_controle = ['India','Brazil', 'Vietnam','Afghanistan']
analisados = afdb + grupo_controle
print (analisados)
```

```python

sel_paises = mi.loc[mi['country'].isin(analisados)]
dados1 = sel_paises.set_index('country')
#dados1 se refere à primeira variável estudada (mortalidade infantil)

sel_paises = hiv.loc[mi['country'].isin(analisados)]
dados2 = sel_paises.set_index('country')
#dados2 se refere à segunda variável estudada (taxa de adultos(15 a 49 anos) 
#porém não há dados suficientes para todos os países estudados)
dados1.loc[:,"AfDB"] = "Não"
dados1.loc[afdb,"AfDB"] = "Sim"
dados1
```

```python
# os anos de 1996 e 2006 foram marcantes pois representaram, respectivamente o quase colapso e um sinal de recuperação do Banco

dados1_anos = dados1.loc[:,[1992,1996,2000,2003,2006,2010,2014,2018]]
dados1_anos.loc[:,"AfDB"] = "Não"
dados1_anos.loc[afdb,"AfDB"] = "Sim"
dados1_anos
```

```python
growth = pd.read_excel('gdp_total_yearly_growth.xlsx')

sele_paises = growth.loc[growth['country'].isin(analisados)]
dados3 = sele_paises.set_index('country')
dados3.loc[:,"AfDB"] = "Não"
dados3.loc[afdb,"AfDB"] = "Sim"
dados3_anos = dados3.loc[:,[1992,1996,2000,2003,2006,2010,2014,2018]]
dados3_anos.loc[:,"AfDB"] = "Não"
dados3_anos.loc[afdb,"AfDB"] = "Sim"
dados3_anos
```
